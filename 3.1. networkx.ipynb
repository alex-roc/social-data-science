# Introducci√≥n a NetworkX: An√°lisis de Redes y Grafos

## ¬øQu√© es NetworkX?

**NetworkX** es una biblioteca de Python para crear, manipular y estudiar redes complejas (grafos). Es perfecta para analizar relaciones entre elementos como:

- Redes sociales (amistades en Facebook)
- Redes de transporte (rutas de metro)
- Redes biol√≥gicas (interacciones entre prote√≠nas)
- Internet y p√°ginas web
- Relaciones comerciales entre empresas

## Conceptos B√°sicos de Grafos

Un **grafo** est√° formado por:

1. **Nodos (Vertices)**: Los elementos individuales (personas, ciudades, p√°ginas web)
2. **Aristas (Edges)**: Las conexiones entre nodos (amistad, carretera, enlace)

### Tipos de Grafos:
- **Dirigidos**: Las conexiones tienen direcci√≥n (A ‚Üí B)
- **No dirigidos**: Las conexiones son bidireccionales (A ‚Üî B)
- **Ponderados**: Las conexiones tienen peso/valor (distancia, intensidad)

En este ejercicio aprenderemos a:
- Crear diferentes tipos de grafos
- Agregar y quitar nodos y aristas
- Analizar propiedades de la red
- Visualizar grafos de manera efectiva

¬°Comencemos con una red social simple! üë•

```python
# Importar las bibliotecas necesarias
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Configurar el estilo de los gr√°ficos
plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 8)

# Verificar la versi√≥n de NetworkX
print(f"Versi√≥n de NetworkX: {nx.__version__}")
print("¬°NetworkX importado correctamente! üï∏Ô∏è")
print()

# Informaci√≥n b√°sica sobre NetworkX
print("üìö NetworkX puede crear grafos:")
print("- Graph(): Grafo no dirigido")
print("- DiGraph(): Grafo dirigido") 
print("- MultiGraph(): Grafo con m√∫ltiples aristas entre nodos")
print("- MultiDiGraph(): Grafo dirigido con m√∫ltiples aristas")
```

## 1. Crear Grafos B√°sicos

Empezemos creando diferentes tipos de grafos con datos sencillos de una red social universitaria.

```python
# 1.1 GRAFO NO DIRIGIDO (amistades bidireccionales)
print("üë• Creando una red social universitaria (grafo no dirigido)")

# Crear un grafo vac√≠o
G = nx.Graph()

# Agregar nodos (estudiantes)
estudiantes = ["Ana", "Bob", "Carlos", "Diana", "Eva", "Frank"]
G.add_nodes_from(estudiantes)

print(f"Nodos en el grafo: {list(G.nodes())}")
print(f"N√∫mero de nodos: {G.number_of_nodes()}")
print(f"N√∫mero de aristas: {G.number_of_edges()}")
print()

# Agregar aristas (amistades)
amistades = [
    ("Ana", "Bob"),
    ("Ana", "Carlos"), 
    ("Bob", "Diana"),
    ("Carlos", "Diana"),
    ("Diana", "Eva"),
    ("Eva", "Frank"),
    ("Carlos", "Frank")
]

G.add_edges_from(amistades)

print("Despu√©s de agregar amistades:")
print(f"N√∫mero de aristas: {G.number_of_edges()}")
print(f"Aristas: {list(G.edges())}")
print()

# Informaci√≥n b√°sica del grafo
print("üìä Informaci√≥n del grafo:")
print(f"¬øEst√° conectado?: {nx.is_connected(G)}")
print(f"Densidad: {nx.density(G):.3f}")  # Qu√© tan conectado est√° (0-1)
```

```python
# 1.2 GRAFO DIRIGIDO (followers en redes sociales)
print("üì± Creando una red de seguidores (grafo dirigido)")

# Crear un grafo dirigido
D = nx.DiGraph()

# Los mismos estudiantes
D.add_nodes_from(estudiantes)

# Relaciones de "seguir" (A sigue a B)
seguimientos = [
    ("Ana", "Bob"),      # Ana sigue a Bob
    ("Bob", "Ana"),      # Bob sigue a Ana (rec√≠proco)
    ("Carlos", "Ana"),   # Carlos sigue a Ana
    ("Diana", "Bob"),    # Diana sigue a Bob
    ("Diana", "Carlos"), # Diana sigue a Carlos
    ("Eva", "Diana"),    # Eva sigue a Diana
    ("Frank", "Eva"),    # Frank sigue a Eva
    ("Frank", "Ana")     # Frank sigue a Ana
]

D.add_edges_from(seguimientos)

print(f"Aristas dirigidas: {list(D.edges())}")
print()

# Comparar con el grafo no dirigido
print("üîÑ Comparaci√≥n dirigido vs no dirigido:")
print(f"Grafo no dirigido - Aristas: {G.number_of_edges()}")
print(f"Grafo dirigido - Aristas: {D.number_of_edges()}")
print()

# An√°lisis espec√≠fico de grafos dirigidos
print("üìà An√°lisis del grafo dirigido:")
for nodo in ["Ana", "Bob", "Diana"]:
    in_degree = D.in_degree(nodo)   # Cu√°ntos lo siguen
    out_degree = D.out_degree(nodo) # A cu√°ntos sigue
    print(f"{nodo}: {in_degree} seguidores, sigue a {out_degree}")
```

```python
# 1.3 GRAFO PONDERADO (intensidad de amistad)
print("üí™ Creando una red con pesos (intensidad de amistad)")

# Crear un grafo ponderado
W = nx.Graph()

# Agregar aristas con pesos (intensidad de amistad del 1 al 10)
amistades_ponderadas = [
    ("Ana", "Bob", {"peso": 8}),      # Muy buenos amigos
    ("Ana", "Carlos", {"peso": 6}),   # Buenos amigos
    ("Bob", "Diana", {"peso": 9}),    # Mejores amigos
    ("Carlos", "Diana", {"peso": 7}), # Buenos amigos
    ("Diana", "Eva", {"peso": 5}),    # Amigos normales
    ("Eva", "Frank", {"peso": 4}),    # Conocidos
    ("Carlos", "Frank", {"peso": 3})  # Apenas se conocen
]

W.add_edges_from(amistades_ponderadas)

print("Aristas con pesos:")
for u, v, data in W.edges(data=True):
    print(f"{u} - {v}: intensidad {data['peso']}")
print()

# Estad√≠sticas de los pesos
pesos = [data['peso'] for u, v, data in W.edges(data=True)]
print(f"üìä Estad√≠sticas de intensidad:")
print(f"Promedio: {np.mean(pesos):.1f}")
print(f"M√≠nimo: {np.min(pesos)}")
print(f"M√°ximo: {np.max(pesos)}")
print()

# Encontrar la amistad m√°s fuerte
amistad_fuerte = max(W.edges(data=True), key=lambda x: x[2]['peso'])
print(f"üèÜ Amistad m√°s fuerte: {amistad_fuerte[0]} - {amistad_fuerte[1]} (intensidad: {amistad_fuerte[2]['peso']})")
```

## 2. Manipulaci√≥n de Grafos

Aprende a agregar, quitar y modificar nodos y aristas din√°micamente.

```python
# 2.1 AGREGAR Y QUITAR NODOS
print("‚ûï Manipulando nodos del grafo")

# Trabajaremos con una copia del grafo original
G_modificado = G.copy()

print(f"Grafo original: {G_modificado.number_of_nodes()} nodos, {G_modificado.number_of_edges()} aristas")
print(f"Nodos: {list(G_modificado.nodes())}")
print()

# Agregar nuevos estudiantes
nuevos_estudiantes = ["Grace", "Hugo"]
G_modificado.add_nodes_from(nuevos_estudiantes)

print(f"Despu√©s de agregar estudiantes: {G_modificado.number_of_nodes()} nodos")
print(f"Nuevos nodos: {list(G_modificado.nodes())}")
print()

# Agregar nuevas amistades
nuevas_amistades = [
    ("Grace", "Ana"),
    ("Hugo", "Frank"),
    ("Grace", "Diana")
]
G_modificado.add_edges_from(nuevas_amistades)

print(f"Despu√©s de agregar amistades: {G_modificado.number_of_edges()} aristas")
print()

# Quitar un nodo (y todas sus conexiones)
print("‚ùå Quitando un nodo:")
print(f"Antes de quitar 'Carlos': {list(G_modificado.neighbors('Carlos'))}")
G_modificado.remove_node("Carlos")
print(f"Despu√©s de quitar 'Carlos': {G_modificado.number_of_nodes()} nodos, {G_modificado.number_of_edges()} aristas")
print(f"Nodos restantes: {list(G_modificado.nodes())}")
```

```python
# 2.2 AGREGAR ATRIBUTOS A NODOS Y ARISTAS
print("üè∑Ô∏è Agregando atributos a nodos y aristas")

# Crear un nuevo grafo para este ejemplo
G_atributos = nx.Graph()

# Agregar nodos con atributos (informaci√≥n de estudiantes)
estudiantes_info = [
    ("Ana", {"edad": 20, "carrera": "Ingenier√≠a", "a√±o": 2}),
    ("Bob", {"edad": 19, "carrera": "Medicina", "a√±o": 1}),
    ("Diana", {"edad": 21, "carrera": "Psicolog√≠a", "a√±o": 3}),
    ("Eva", {"edad": 22, "carrera": "Arte", "a√±o": 4})
]

G_atributos.add_nodes_from(estudiantes_info)

# Agregar aristas con atributos (tipo de relaci√≥n y tiempo de amistad)
amistades_info = [
    ("Ana", "Bob", {"tipo": "compa√±eros", "a√±os_amistad": 2}),
    ("Bob", "Diana", {"tipo": "mejores_amigos", "a√±os_amistad": 3}),
    ("Diana", "Eva", {"tipo": "conocidos", "a√±os_amistad": 1})
]

G_atributos.add_edges_from(amistades_info)

print("üë§ Informaci√≥n de nodos:")
for nodo, data in G_atributos.nodes(data=True):
    print(f"{nodo}: {data}")
print()

print("üîó Informaci√≥n de aristas:")
for u, v, data in G_atributos.edges(data=True):
    print(f"{u} - {v}: {data}")
print()

# Filtrar por atributos
print("üéØ Filtros por atributos:")
estudiantes_mayores = [nodo for nodo, data in G_atributos.nodes(data=True) if data['edad'] >= 21]
print(f"Estudiantes de 21+ a√±os: {estudiantes_mayores}")

mejores_amigos = [arista for u, v, data in G_atributos.edges(data=True) if data['tipo'] == 'mejores_amigos']
print(f"Mejores amigos: {mejores_amigos}")
```

## 3. An√°lisis de Grafos

Ahora exploraremos m√©tricas importantes para entender la estructura de la red.

```python
# 3.1 GRADO DE NODOS (n√∫mero de conexiones)
print("üìä An√°lisis de grados en la red social")

# Usar el grafo original G
print("Grado de cada estudiante (n√∫mero de amigos):")
for nodo in G.nodes():
    grado = G.degree(nodo)
    amigos = list(G.neighbors(nodo))
    print(f"{nodo}: {grado} amigos -> {amigos}")
print()

# Estad√≠sticas de grados
grados = dict(G.degree())
print("üìà Estad√≠sticas de grados:")
print(f"Promedio: {np.mean(list(grados.values())):.1f}")
print(f"M√°ximo: {max(grados.values())}")
print(f"M√≠nimo: {min(grados.values())}")
print()

# Encontrar el nodo m√°s popular
nodo_popular = max(grados, key=grados.get)
print(f"üåü Estudiante m√°s popular: {nodo_popular} con {grados[nodo_popular]} amigos")

# Crear un DataFrame para mejor visualizaci√≥n
df_grados = pd.DataFrame(list(grados.items()), columns=['Estudiante', 'N√∫mero_Amigos'])
df_grados = df_grados.sort_values('N√∫mero_Amigos', ascending=False)
print("\nüìã Ranking de popularidad:")
print(df_grados)
```

```python
# 3.2 CAMINOS Y DISTANCIAS
print("üõ§Ô∏è An√°lisis de caminos en la red")

# Camino m√°s corto entre dos nodos
try:
    camino = nx.shortest_path(G, "Ana", "Frank")
    longitud = nx.shortest_path_length(G, "Ana", "Frank")
    print(f"Camino m√°s corto de Ana a Frank: {' -> '.join(camino)}")
    print(f"Longitud del camino: {longitud} pasos")
except nx.NetworkXNoPath:
    print("No hay camino entre Ana y Frank")
print()

# Distancias entre todos los pares de nodos
print("üìè Matriz de distancias (caminos m√°s cortos):")
distancias = dict(nx.all_pairs_shortest_path_length(G))

# Crear una matriz de distancias
nodos = list(G.nodes())
matriz_dist = []
for i, nodo1 in enumerate(nodos):
    fila = []
    for j, nodo2 in enumerate(nodos):
        if nodo1 in distancias and nodo2 in distancias[nodo1]:
            fila.append(distancias[nodo1][nodo2])
        else:
            fila.append(np.inf)  # No hay camino
    matriz_dist.append(fila)

df_distancias = pd.DataFrame(matriz_dist, index=nodos, columns=nodos)
print(df_distancias)
print()

# Di√°metro de la red (distancia m√°s larga)
if nx.is_connected(G):
    diametro = nx.diameter(G)
    print(f"üî¥ Di√°metro de la red: {diametro} (m√°xima distancia entre cualquier par de nodos)")
else:
    print("üî¥ La red no est√° completamente conectada")
```

```python
# 3.3 MEDIDAS DE CENTRALIDAD (importancia de nodos)
print("‚≠ê Medidas de centralidad - ¬øQui√©n es m√°s importante en la red?")

# Centralidad de grado (ya la vimos)
degree_centrality = nx.degree_centrality(G)

# Centralidad de intermediaci√≥n (betweenness)
betweenness_centrality = nx.betweenness_centrality(G)

# Centralidad de cercan√≠a (closeness)
closeness_centrality = nx.closeness_centrality(G)

# Centralidad de vector propio (eigenvector)
eigenvector_centrality = nx.eigenvector_centrality(G)

print("üìä Medidas de centralidad:")
print("Nodo\t\tGrado\t\tIntermed.\tCercan√≠a\tEigenvector")
print("-" * 65)

for nodo in G.nodes():
    print(f"{nodo:<10}\t{degree_centrality[nodo]:.3f}\t\t{betweenness_centrality[nodo]:.3f}\t\t{closeness_centrality[nodo]:.3f}\t\t{eigenvector_centrality[nodo]:.3f}")

print()
print("üìù Interpretaci√≥n:")
print("- Grado: Popularidad directa (m√°s conexiones)")
print("- Intermediaci√≥n: Importancia como puente entre otros")
print("- Cercan√≠a: Qu√© tan cerca est√° de todos los dem√°s")
print("- Eigenvector: Importancia basada en la importancia de sus conexiones")
print()

# Encontrar el nodo m√°s central en cada m√©trica
print("üèÜ Nodos m√°s centrales:")
print(f"Mayor grado: {max(degree_centrality, key=degree_centrality.get)}")
print(f"Mayor intermediaci√≥n: {max(betweenness_centrality, key=betweenness_centrality.get)}")
print(f"Mayor cercan√≠a: {max(closeness_centrality, key=closeness_centrality.get)}")
print(f"Mayor eigenvector: {max(eigenvector_centrality, key=eigenvector_centrality.get)}")
```

## 4. Visualizaci√≥n de Grafos

La visualizaci√≥n nos ayuda a entender mejor la estructura de la red.

```python
# 4.1 VISUALIZACI√ìN B√ÅSICA
print("üé® Creando visualizaciones de la red social")

# Configurar el layout
plt.figure(figsize=(15, 5))

# Tres visualizaciones diferentes
layouts = [
    ("spring", nx.spring_layout(G, seed=42)),
    ("circular", nx.circular_layout(G)),
    ("random", nx.random_layout(G, seed=42))
]

for i, (nombre, pos) in enumerate(layouts, 1):
    plt.subplot(1, 3, i)
    
    # Dibujar el grafo
    nx.draw(G, pos, 
            with_labels=True, 
            node_color='lightblue',
            node_size=1000,
            font_size=10,
            font_weight='bold',
            edge_color='gray')
    
    plt.title(f'Layout: {nombre}')

plt.tight_layout()
plt.show()

print("üìù Diferentes layouts muestran la misma red de maneras distintas")
print("- Spring: Algoritmo de fuerzas (nodos conectados se atraen)")
print("- Circular: Nodos dispuestos en c√≠rculo")
print("- Random: Posiciones aleatorias")
```

```python
# 4.2 VISUALIZACI√ìN AVANZADA CON COLORES Y TAMA√ëOS
print("üåà Visualizaci√≥n avanzada basada en m√©tricas")

plt.figure(figsize=(12, 4))

# Preparar datos para visualizaci√≥n
pos = nx.spring_layout(G, seed=42)
grados = dict(G.degree())
centralidad = nx.betweenness_centrality(G)

# Visualizaci√≥n 1: Tama√±o por grado
plt.subplot(1, 2, 1)
node_sizes = [grados[nodo] * 300 for nodo in G.nodes()]
nx.draw(G, pos,
        with_labels=True,
        node_color='lightcoral',
        node_size=node_sizes,
        font_size=9,
        font_weight='bold',
        edge_color='gray')
plt.title('Tama√±o = N√∫mero de amigos')

# Visualizaci√≥n 2: Color por centralidad de intermediaci√≥n
plt.subplot(1, 2, 2)
node_colors = [centralidad[nodo] for nodo in G.nodes()]
nodes = nx.draw_networkx_nodes(G, pos,
                              node_color=node_colors,
                              node_size=800,
                              cmap='viridis')
nx.draw_networkx_labels(G, pos, font_size=9, font_weight='bold')
nx.draw_networkx_edges(G, pos, edge_color='gray')
plt.colorbar(nodes, label='Centralidad de intermediaci√≥n')
plt.title('Color = Importancia como puente')

plt.tight_layout()
plt.show()

print("üí° Interpretaci√≥n:")
print("- Nodos m√°s grandes = m√°s conexiones directas")
print("- Colores m√°s claros = mayor importancia como intermediarios")
```

Ahora es momento de que practiques con los conceptos de NetworkX. Aqu√≠ tienes algunos ejercicios:

```python
# EJERCICIO 1: Red de transporte urbano
print("üöá EJERCICIO 1 - Crear una red de transporte")

# Crear un grafo de estaciones de metro
metro = nx.Graph()

# Agregar estaciones (nodos)
estaciones = ["Centro", "Universidad", "Hospital", "Aeropuerto", "Estadio", "Mall", "Puerto"]
metro.add_nodes_from(estaciones)

# Agregar conexiones entre estaciones con tiempo de viaje (minutos)
rutas = [
    ("Centro", "Universidad", {"tiempo": 5}),
    ("Centro", "Hospital", {"tiempo": 8}),
    ("Universidad", "Mall", {"tiempo": 12}),
    ("Hospital", "Aeropuerto", {"tiempo": 15}),
    ("Mall", "Estadio", {"tiempo": 7}),
    ("Estadio", "Puerto", {"tiempo": 10}),
    ("Centro", "Puerto", {"tiempo": 20})  # L√≠nea directa
]

metro.add_edges_from(rutas)

print(f"Red de metro: {metro.number_of_nodes()} estaciones, {metro.number_of_edges()} rutas")
print()

# An√°lisis de la red de metro
print("üìä An√°lisis de la red de metro:")

# ¬øCu√°l es la estaci√≥n m√°s conectada?
grados_metro = dict(metro.degree())
estacion_central = max(grados_metro, key=grados_metro.get)
print(f"Estaci√≥n m√°s conectada: {estacion_central} ({grados_metro[estacion_central]} conexiones)")

# Ruta m√°s corta de Universidad a Aeropuerto
ruta_uni_aero = nx.shortest_path(metro, "Universidad", "Aeropuerto")
print(f"Ruta m√°s corta Universidad ‚Üí Aeropuerto: {' ‚Üí '.join(ruta_uni_aero)}")

# Tiempo total de la ruta (suma de pesos)
tiempo_total = nx.shortest_path_length(metro, "Universidad", "Aeropuerto", weight="tiempo")
print(f"Tiempo total del viaje: {tiempo_total} minutos")
```
